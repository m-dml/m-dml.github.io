```python
import numpy as np
from matplotlib import pyplot as plt
x = np.load('Z500_T850_example_preds_N5_dt24h.npy')
```

```python
n = x.shape[0]
x.shape
```

```python
names = ['z500', 't850']
for i in range(2):
    plt.figure(figsize=(10, 20))
    for j in range(n):
        ax = plt.subplot(1, n, j + 1)
        plt.imshow(x[j,i,:,:])
        plt.axis('off')
        plt.savefig(names[i] + '.png', bbox_inches='tight', transparent=True)
```
